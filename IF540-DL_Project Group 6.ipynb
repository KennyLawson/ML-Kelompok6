```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bc = pd.read_csv("breast-cancer-wisconsin-data_data.csv")
bc.head()
```

## DATA VISUALIZATION

```python
bc.info()
```

```python
bc.describe()
```

```python
bc['diagnosis'].value_counts()
```

```python
labels =['Benign','Malignant']
sizes = bc['diagnosis'].value_counts() 
colors = ['pink', 'red']
explode = (0, 0) 

plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', startangle=90)

plt.axis('equal') 
plt.title('Diagnosis Distribution')

plt.show()
```

## DATA PREPROCESSING

##### Drop Invalid column

```python
bc = bc.drop(columns=['Unnamed: 32'], errors='ignore')
```

##### Handling Missing Value

```python
bc.isnull().sum()
```

```python
sns.heatmap(bc.isnull(), cbar=False, cmap="plasma") 
sns.set(rc = {'figure.figsize': (20,10)}) 
plt.title('Missing Value') 
plt.show()
```

```python
bc.boxplot()
plt.show()
```

#### Encoding

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
bc['diagnosis'] = le.fit_transform(bc['diagnosis'])
```

```python
correlation_index = bc.corrwith(bc['diagnosis'])
```

```python
sns.heatmap(bc.corr(), cmap='jet')
```

## MODELLING

```python
X = bc.drop(['diagnosis'], axis=1)  
y = bc['diagnosis']  
```

#### Data Splitting

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=45)
```

#### Data Training

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
```

```python
score = model.score(X_test, y_test)
print("Accuracy:", score)
```

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(X_train, y_train)
svm_score = svm.score(X_test, y_test)
```

```python
svm = SVC()
svm.fit(X_train, y_train)
```

```python
svm_score = svm.score(X_test, y_test)
print('Akurasi SVM:', svm_score)
```

```python
y_pred = model.predict(X_test)
```

### Evaluation

### Confussion Matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(10, 6))

sns.heatmap(
    cm, annot=True, fmt='d', cmap='magma', linewidths=0.4, square=True, cbar=True,
    xticklabels=["Benign", "Malignant"],
    yticklabels=["Benign", "Malignant"]
)

plt.xlabel('Predicted', fontsize=14, fontweight='bold')
plt.ylabel('Actual', fontsize=14, fontweight='bold')
plt.title('Confusion Matrix', fontsize=16, fontweight='bold')
plt.yticks(rotation=360)

plt.show()
```
